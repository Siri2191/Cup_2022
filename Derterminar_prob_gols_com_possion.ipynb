## calculando probabilidade de gol de um time com poisson:
---

```python
mean_ecuador = gols_ecuador.mean()
list_gols = [0,1,2,3,4,5,6,7,8]
list_prob_ecuador = []

from scipy.stats import poisson

for i in list_gols:
  prob = poisson.pmf(i,mean_ecuador)
  list_prob_ecuador.append(prob) 

list_prob_ecuador
```

## Juntando em um Script:
---

```python
# selecionar time:

time_a = input()
time_b = input()

# tratando a entrada:
time_a = time_a.capitalize()
time_b = time_b.capitalize()

# selecionar jogos:

# time_a :
casa_a = df[df['time_casa']== time_a]
fora_a = df[df['time_visitante']== time_a]
a = casa_a.merge(fora_a,how='outer')

#time_b:
casa_b = df[df['time_casa']== time_b]
fora_b = df[df['time_visitante']== time_b]
b = casa_b.merge(fora_b,how='outer')

# selecionar gols do jogos:

# time a:
casa_a= a.loc[a['time_casa'] == time_a ]['placar_casa']
fora_a = a.loc[a['time_visitante'] == time_a ]['placar_visitante']
gols_a = casa_a.append(fora_a)

# time b:
casa_b = b.loc[b['time_casa'] == time_b ]['placar_casa']
fora_b = b.loc[b['time_visitante'] == time_b ]['placar_visitante']
gols_b = casa_a.append(fora_b)

# calculo das probabilidade: 

mean_a = gols_a.mean()
mean_b = gols_b.mean()
list_gols = [0,1,2,3,4,5]
prob_a = []
prob_b = []

for i in list_gols:
  pb_a = poisson.pmf(i,mean_a)
  pb_b = poisson.pmf(i,mean_b) 
  prob_a.append(pb_a)
  prob_b.append(pb_b)

# probabilidades das possibilidades de placar: 
df_possibilidade = pd.DataFrame(columns = ['time_a',
                              'gol_a',
                              'gol_b',
                              'time_b',
                              'probabilidade_poisson' ])

for i in range(0,len(prob_a)):
  for j in range(0,len(prob_b)):
    df_possibilidade = df_possibilidade.append({'time_a':time_a,
                              'gol_a': i,
                              'gol_b': j,
                              'time_b': time_b,
                              'probabilidade_poisson': (prob_a[i]*prob_b[j])*100 }, ignore_index =True)

df_possibilidade.sort_values('probabilidade_poisson',ascending = False)
```
